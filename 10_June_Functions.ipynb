3. What is the purpose of the return statement in a function? Can a function have multiple return statements? Explain with an example.
Ans. The purpose of the return statement in a function is to specify the value or values that the function should produce as its result.
When a function is called and the return statement is encountered, the function's execution is immediately terminated, and the value provided after the return keyword is sent back as the output of the function.

Yes, a function can have multiple return statements. 
def check_number(number):
    if number > 0:
        return "Positive"
    elif number < 0:
        return "Negative"
    else:
        return "Zero"
result1 = check_number(5)
result2 = check_number(-3)
result3 = check_number(0)

print(result1) # Output: "Positive"
print(result2) # Output: "Negative"
print(result3) # Output: "Zero"

4. What are lambda functions in Python? How are they different from regular functions? Provide an example where a lambda function can be useful.
Ans. Lambda functions in Python are a way to create small, anonymous, and inline functions. 
Difference between lambda and regular functions are:-
1 Lambda functions use the lambda keyword, whereas regular functions are defined using the def keyword.
2 Lambda functions are usually one-liners that focus on performing simple operations. Regular functions can be multi-line and more complex.
3 Lambda functions don't have a name, while regular functions require a name for reference and later use.
# Regular function to add two numbers
def add(a, b):
    return a + b

# Equivalent lambda function to add two numbers
add_lambda = lambda a, b: a + b

5. How does the concept of "scope" apply to functions in Python? Explain the difference between local scope and global scope.
Ans. scope refers to the visibility and accessibility of variables within different parts of a Python program. 
1 Local Scope:
Variables defined inside a function are considered to be in a local scope. These variables can only be accessed and used within the function where they are defined.
eg - 
def abc():
    x = 10  # This variable `x` has local scope within the function
    print(x)

abc() # Output: 10
2 Global Scope:
Variables defined outside any function, at the top level of a script or module, have a global scope. These variables can be accessed and modified from any part of the code, including inside functions.
eg-
y = 5  # This variable `y` has global scope
def abc():
    global y  # Using the `global` keyword to access the global variable `y`
    y += 1   # Modifying the global variable `y`
    print(y)

abc() # Output: 6

print(y)  # The modified value of the global variable `y` is visible outside the function  # Output: 6

6. How can you use the "return" statement in a Python function to return multiple values?
Ans. To return multiple values from a function, you can use various approaches. Here are some common methods:
1.Returning a tuple:
def multiple_values():
    value1 = 10
    value2 = "Hello"
    value3 = [1, 2, 3]
    return value1, value2, value3

result1, result2, result3 = multiple_values()
print(result1)  # Output: 10
print(result2)  # Output: Hello
print(result3)  # Output: [1, 2, 3]

2.Using a list or dictionary:
Using a list:
def multiple_values_list():
    value1 = 10
    value2 = "Hello"
    value3 = [1, 2, 3]
    return [value1, value2, value3]

result_list = multiple_values_list()
print(result_list[0])  # Output: 10
print(result_list[1])  # Output: Hello
print(result_list[2])  # Output: [1, 2, 3]

Using a dictionary:
def multiple_values_dict():
    value1 = 10
    value2 = "Hello"
    value3 = [1, 2, 3]
    return {'value1': value1, 'value2': value2, 'value3': value3}

result_dict = multiple_values_dict()
print(result_dict['value1'])  # Output: 10
print(result_dict['value2'])  # Output: Hello
print(result_dict['value3'])  # Output: [1, 2, 3]

7. What is the difference between the "pass by value" and "pass by reference" concepts when it comes to function arguments in Python?
Ans. 1. Pass by value:It means that when you pass an argument to a function, a copy of the argument's value is made, and the function works with this copy rather than the original data. 
def modify_value(x):
    x += 1
    return x

num = 10
modified_num = modify_value(num)
print(num)          # Output: 10 (unchanged)
print(modified_num) # Output: 11
 
2.Pass by reference: It means that when you pass an argument to a function, the function operates directly on the original data rather than a copy. 
def modify_list(lst):
    lst.append(4)

my_list = [1, 2, 3]
modify_list(my_list)
print(my_list) # Output: [1, 2, 3, 4] (modified)
