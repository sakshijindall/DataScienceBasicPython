5. How many global scopes are there in a Python program? How many local scopes?
Ans. Python program has one global scope that remains accessible throughout the program whereas each function has its own local scope that is created when the function is called and destroyed when the execution is completed.

6. What happens to variables in a local scope when the function call returns?
Ans. When the function call returns in python the local scope is destroyed and hence the variable within that local scope does not exists. This proceas ia known as variable cleanup or garbage collection. 
If we try to call that variable it will give us 'NameError'.

Ans. By returning values from functions, you can create reusable and modular code that performs specific computations and produces results that can be utilized in different parts of your program.

it is not possible to have return value directly in an expression. The return statement itself is a statement, not an expression. It does not produce a value that can be directly used within an expression or assigned to a variable in the middle of an expression.

8. If a function does not have a return statement, what is the return value of a call to that function?
Ans. If a function does not have a return statement or if it reaches the end of the function without including a return statement, the function will return 'None'in Python.

None is a special value in Python that represents the absence of a value or the lack of a return value. It is commonly used to indicate that a function does not return anything meaningful or when the result is not applicable.

14. What is the purpose of the try clause? What is the purpose of the except clause?
Ans. The purpose of the try clause is to have the block of code that may raise an exception. The code within the try block is executed and if any exception occurs it will immediately moves to the except block.
The purpose of the except clause is define how to handle specific types of exception that might occur with the corresponding try block.
